```python
#Importing Libraries
import pandas as pd
import re
import numpy as np
```

```python
#Loading Dataset
df = pd.read_csv('/Users/saarth2712/Downloads/Uncleaned_DS_jobs.csv')
df.info()
```

```python
#Cleaning
df.dropna
df.drop_duplicates()
```

```python
#Salary Estimate
df['Salary Estimate'].head()    
```

```python
#Converting Salary Estimate to Integer
def convert_salary_estimate(salary) :
    salary = re.sub(r'[^0-9\-K]', '', salary) 
    
    if '-' in salary:
        min_salary, max_salary = salary.split('-')
        
        min_salary = int(min_salary.replace('K', '')) * 1000
        max_salary = int(max_salary.replace('K', '')) * 1000
        
        return (min_salary + max_salary) // 2

df['Average Salary'] = df['Salary Estimate'].apply(convert_salary_estimate)   
```

```python
df['Average Salary'].head()
```

```python
#Job Description
print(df['Job Description'].apply(type).unique())
df['Job Description'] = df['Job Description'].astype(str)
df['Job Description'].head()
```

```python
#Cleaning Job Description
def clean_text(text):
    if isinstance(text, str):  # Ensure the value is a string
        # Remove extra newlines and convert them to a single space
        cleaned_text = text.replace("\n\n", " ").replace("\n", " ").strip()
        return cleaned_text
    return ""

df['Job Description'] = df['Job Description'].apply(clean_text)
df['Job Description']
```

```python
#Company Name
df['Company Name']
```

```python
#Cleaning Company Name
df['Company Name'].dtype
def clean_company_name(name) :
    name = name.replace("\n", " ").strip()
    name = re.sub(r'[^a-zA-Z\s]', '', name)
    name = re.sub(r'\d+', '', name)
    name = name.strip()
    
    return name

df['Company Name'] = df['Company Name'].apply(clean_company_name)
df['Company Name']
```

```python
#Dropping Negative Values
df = df[~df.select_dtypes(include=['number']).lt(0).any(axis=1)]
df
```

```python
#Competitors
df['Competitors']
```

```python
#Revenue
df['Revenue'] = df['Revenue'].str.replace(r'\s\(USD\)', '', regex=True)
df['Revenue']
```

```python
#Competitors
df['Competitors'] = df['Competitors'].replace('-1', np.nan)
df['Competitors']
```

```python
#Creating State Column
df['State'] = df['Location'].str[-2:]
df
```
